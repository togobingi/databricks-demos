```python
CATALOG = dbutils.widgets.get("catalog")
SCHEMA = dbutils.widgets.get("schema")
TABLE = "trading_events"
```

```python


# read in target table
df = spark.readStream.table(f"{CATALOG}.{SCHEMA}.{TABLE}")

checkpoint_path = f"/Volumes/{CATALOG}/{SCHEMA}/streaming_checkpoints"
output_table = f"{CATALOG}.{SCHEMA}.trading_events_processed"

# write to a new delta table
df.writeStream.format("delta").option("checkpointLocation", checkpoint_path).trigger(availableNow=True).table(f"{output_table}")
```
